```python
#Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import copy
import warnings

plt.rc('font', size=12)
warnings.filterwarnings('ignore')
```

```python
#Helper Methods
def plotWaveForm(df, colName, startTime=0, endTime=None):
    if endTime != None:
        waves = df.iloc[startTime:endTime][colName]
        plt.figure(figsize=(16,7))
        plt.plot(waves)
        plt.xlabel("Time")
        plt.ylabel("PWV")
        plt.title("Pulse Wave Signal")
    else:
        plt.figure(figsize=(16,7))
        plt.plot(df[colName])
        plt.xlabel("Time")
        plt.ylabel("PWV")
        plt.title("Pulse Wave Signal")
```

```python
#Load patient data
x1 = pd.read_csv('Data/Raw Data/Multiple Cath/X001.txt', sep=" ")
x1
```

```python
#drop empty last column
x1 = x1.iloc[:,:-1]
x1
```

```python
plotWaveForm(df=x1, colName='AO', startTime=0, endTime=None)
```

```python
plotWaveForm(df=x1, colName='A0', startTime=0, endTime=2000)
```
